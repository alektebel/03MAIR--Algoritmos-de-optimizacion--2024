#Respuesta
- ¿Como represento el espacio de soluciones?

El espacio de solucines es la generación de 15 franjas horarias junto con la asignación respectiva de profesores. Es decir, una matriz 35x10x3 con unos y ceros, que representa todas las posibles combinaciones de jurados. Las dimensiones representan respectivamente 35 posibles franjas horarias, para 10 posibles profesores, realizando 10 posibles roles.


- ¿Cual es la función objetivo?

Puesto que tenemos la posibilidad de explotar los profesores que tienen la posibilidad de realizar más roles, se propone tomar como función objetivo la entropía de las frecuencias de cada profesor. De esta forma, si un profesor tiene demasiada frecuencia, su entropía será muy alta, y por tanto será una solución poco deseable. Por tanto, compararemos soluciones con su entropía, y escogeremos la que menor entropía tenga (es decir, que sea la más equitativa en frecuencias entre sus profesores)
También valdría la varianza de las frecuencias. Puesto que esta es más sencilla de implementar, utilizaremos esta.

$$f(a_{ijk}) = \sum_j (\sum_{i, k} a_{ijk} - \frac{1}{profesores} \sum_j \sum_{ik} a_{ijk} )^2$$

donde $a_ijk$ son las variables a afinar del problema, que explico un poco más adelante.

- ¿Como implemento las restricciones?

Esta sería la parte clave del problema, donde tenemos que expresar con una fórmula las restricciones que tienen que cumplir las variables.
Para empezar, tenemos que encontrar un vocal, un S y un V para cada horario, 15 veces distintas. Esto puede expresarse de la siguiente forma:

Sea $H$ una matriz 35x10, con filas de profesores y columnas de franja horaria en día.
Sea $R$ una matriz 10x3, con filas de profesores y 3 columnas con los roles de cada profesor qeu puede desempeñar, informada a unos y ceros.

Sea $A$ una matriz 28x10, con filas de profesores y columnas de franja horaria en día, que representa las asignaciones que se realizan para cada horario.
* Sus sumas por fila representan los turnos que tiene cada profesor
* Sus sumas por columna representan cuántos profesores hay destinados en cada franja horario.
$A_{ijk} = 1$ si el profesor j está asignado a la franja horaria i para el rol k. En caso contrario, $A_{ijk} = 0$

Claramente, una de las condiciones es que la suma total de los elementos de A sea 15x3=45, y adicionalmente que ningún profesor tome ningún rol que no pueda junto con que ningún profesor tenga un tribunal en un horario que no pueda.

* Ningún profesor puede desempeñar más de un rol simultáneamente:
  * condición 1 > $ \sum_{k} a_{ijk} \le 1 $

* El profesor tiene que poder desempeñar el rol al que ha sido asignado
  * condición 2 > $ a_{ijk} \le r_{jk} \forall i $

Para cada tribunal escogido tenemos tres roles diferentes asignados, y hay en total 15 tribunales.

* En la misma franja de tiempo, dos profesores distintos no deberían estar desempeñando mismos roles.
  * condición 3 > $ \sum_{j} a_{ijk}  \le 1 $

* Deberían haber en total 45 asignaciones distintas de profesores, lo que equivale a 15 tribunales
  * condición 4 > $ \sum_{ijk} a_{ijk} = 45$

* Las asignaciones tienen que estar dentro de los horarios permitidos:
  * condición 5 > $ h_{ij} \geq a_{ji}$

```
import numpy as np
import heapq

#* Ningún profesor puede desempeñar más de un rol simultáneamente:
#  * condición 1 > $ \sum_{k} a_{ijk} \le 1 $
#* El profesor tiene que poder desempeñar el rol al que ha sido asignado
#  * condición 2 > $ a_{ijk} \le r_{jk} \forall i $
#* En la misma franja de tiempo, dos profesores distintos no deberían estar desempeñando mismos roles.
#  * condición 3 > $ \sum_{j} a_{ijk}  \le 1 $
#* Deberían haber en total 45 asignaciones distintas de profesores, lo que equivale a 15 tribunales
#  * condición 4 > $ \sum_{ijk} a_{ijk} = 45$
#* Las asignaciones tienen que estar dentro de los horarios permitidos $$ i = 1, \cdots, 28, j= 1, \cdots, 10 $$
#  * condición 5 > $ h_{ji} \geq a_{ijk}$ para todo k.

def condicion1(A):
  return np.all(A.sum(axis = 2)<= 3)
def condicion2(A, R):
  return np.all(A.max(axis = 1) <= R)

def condicion3(A):
  return np.all(A.sum(axis = 0) <= 1)

def condicion4(A):
  return (sum(A.sum(axis = 0).sum(axis = 0)) == 45)

def condicion5(A, H):
  return np.all(A.max(axis = 2) <= H)

def condiciones(A, H, R):
    """
    Comprobamos todas las condiciones del problema
    """
    return condicion1(A) and condicion2(A, R) and condicion3(A) and condicion4(A) and condicion5(A, H)

def solucion_voraz_inicial(H, R):
    """
    Función que con un horario H y unos roles posibles R, devuelve una solución inicial para las asignaciones de profesores a franjas para la creación de tribunales
    """
    H_temp = H.copy()  # Copy of availability matrix
    numero_profesores = H.shape[0]  # Number of tribunals
    franjas_horarias = H.shape[1]  # Number of professors
    roles = R.shape[1]
    asignacion_roles = np.zeros((franjas_horarias, roles))
    A = np.zeros((numero_profesores, franjas_horarias, roles), dtype=int)  # Assignment matrix
    # Continue until all tribunals have exactly 3 roles filled
    while not condiciones(A, H, R):

        # Iteramos por las franjas horarias
      for franja in range(franjas_horarias):
          # Comprobamos si hay ya un tribunal asignado en este slot.
          if sum(A[:, franja, :].sum(axis = 1)) < 3:
              # Para cada rol, tenemos que escoger un profesor
              for i, rol in enumerate(['P', 'S', 'V']):
                  #En caso de que no haya ningún profesor asignado a este rol:
                  if np.all(A[:, franja, i] == 0):
                      # Encontramos un profesor que esté disponible para este rol, y no haya sido elegido antes.
                      # H_temp[franja, profesor]=1 significa que el profesor está disponible
                      # R.iloc[profesor][rol] = 1 significa que este profesor tiene este rol
                      for profesor in range(numero_profesores):
                          if H_temp[profesor, franja] == 1 and R.iloc[profesor][rol] == 1:
                              # Asignamos este profesor  al slot
                              A[profesor, franja, i] = 1
                              #asignacion_roles[franja,i] = 1
                              H_temp[profesor, franja] = 0  # Informamos al profesor como no disponible


                              if condiciones(A, H, R):
                                return A
                              break

    return A
```

```


import numpy as np
import heapq
import io
import pandas as pd
import random


def vecino_aleatorio(A, H, R, intentos = 1000):
    """Genera una solución vecina aleatoria."""

    # Copiar la solución actual para no modificarla directamente
    nuevo_A = np.copy(A)
    for _ in range(intentos):
      # Elegir dos posiciones aleatorias en la matriz de asignación
      prof1, franja1, rol1 = np.random.randint(0, A.shape[0]), np.random.randint(0, A.shape[1]), np.random.randint(0, A.shape[2])
      prof2, franja2, rol2 = np.random.randint(0, A.shape[0]), np.random.randint(0, A.shape[1]), np.random.randint(0, A.shape[2])

      nuevo_A[prof1, franja1, rol1], nuevo_A[prof2, franja2, rol2] = nuevo_A[prof2, franja2, rol2], nuevo_A[prof1, franja1, rol1]

      # Intercambiar las asignaciones si cumplen las restricciones
      if condiciones(nuevo_A, H, R):
        return nuevo_A
      else:
        #Volvemos a permutarlo de vuelta a lo anterior en caso de que no cumpla la condición
        nuevo_A[prof1, franja1, rol1], nuevo_A[prof2, franja2, rol2] = nuevo_A[prof2, franja2, rol2], nuevo_A[prof1, franja1, rol1]
      return nuevo_A


def busqueda_local(A_inicial, H, R, iteraciones=1000):
  """Búsqueda local con vecinos aleatorios."""
  mejor_A = A_inicial
  mejor_varianza = varianza(mejor_A)

  for _ in range(iteraciones):
      vecino = vecino_aleatorio(mejor_A, H, R)
      if condiciones(vecino, H, R):  #Verificar que la solución vecina cumple las restricciones
          varianza_vecino = varianza(vecino)
          if varianza_vecino < mejor_varianza:
              mejor_A = vecino
              mejor_varianza = varianza_vecino
              print("Se ha encontrado una mejor varianza: ", mejor_varianza)

  return mejor_A


Horarios = H.values
R = roles[["P", "S", "V"]]
A_inicial = solucion_voraz_inicial(Horarios, R)


# Aplicar búsqueda local para mejorar la solución inicial
A_optimizada = busqueda_local(A_inicial, Horarios, R, iteraciones = 10000)

frecuencias_profesores_optimizadas = A_optimizada.sum(axis=1).sum(axis=1)
print(frecuencias_profesores_optimizadas)
print(f"Varianza optimizada: {varianza(A_optimizada)}")
```

Vemos que alcanzamos una solución muy optimizada, que además cumple la condición impuesta. Vemos que la mejora de la varianza de las frecuencias de profesores se detiene en este punto, por lo que es poco probable encontrar una solución mejor.

# Respuesta
Como hemos utilizado heurísticas para resolver este problema, no tenemos una forma consistente para determinar la complejidad, puesto que la busqueda local en este caso no es exhaustiva.

El espacio posible de soluciones, es todas las posibles matrices 35x10x3, restringido a las condiciones impuestas en el enunciado. Estas son 5 condiciones, que reducen el espacio de $2^{1050}$ inicial.

Comenzamos con la condición 4, una de las más restrictivas. Esta, $ \sum_{ijk} a_{ijk} = 45$, reduce el número de soluciones totales a $\binom{1050}{45}$

condición 1 > $ \sum_{k} a_{ijk} \le 1 $
Esta condición nos reduce adicionalmente el espacio posible de soluciones, ya que básicamente quedaría en una matriz de 35 x 10 que puede tener unos y ceros. Por tanto, serían todos los posibles subgrupos de 45 de 35x10, que es:
$$\binom{350}{45}$$


condición 2 > $ a_{ijk} \le r_{jk} \forall i $

Esta condición afecta según haya más profesores con mayor capacidad de roles o menos. En caso de que todos los profesores puedan realizar todos los roles, no afecta al espacio de soluciones. En caso contrario reduciría bastante el espacio de soluciones.


condición 3 > $ \sum_{j} a_{ijk}  \le 1 $

Esto reduce de forma similar el espacio de soluciones a lo que vimos antes, a grupos de 45 sobre 35x3 = 105 posibles elementos. Es decir, podemos reducir el espacio de soluciones realmente a $$\binom{105}{45}$$ gracias a esta condición.
Este número es aproximadamente $2.533×10^{28}$, que es un número extremadamente alto.

Las asignaciones tienen que estar dentro de los horarios permitidos $$ i = 1, \cdots, 28, j= 1, \cdots, 10 $$

 condición 5 > $ h_{ij} \geq a_{ji}$
Esta, al igual que con los roles, depende mucho de la disponibilidad de los horarios, que en este caso no limita mucho el número posible de soluciones.
